```python
# Let's cluster the data based on age and spending score
X = df.loc[:,['Age', 'Spending Score (1-100)']].values
```

```python
from sklearn.cluster import KMeans
```

```python
wcss = []
for k in range(1,11):
    kmeans = KMeans(n_clusters=k, init="k-means++")
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)

plt.figure(figsize = (15,5))
plt.plot(range(1,11),wcss, linewidth = 3, marker = "8", color = "red")
```

```python
kmeans = KMeans(n_clusters = 4)
kmeans.fit(X)
kmeans.predict(X)
```

```python
kmeans.cluster_centers_
```

```python
plt.scatter(X[:,0], X[:,1], c = kmeans.labels_, cmap = "rainbow")
plt.title("Plot between Age and Spending Score")
plt.xlabel("Age")
plt.ylabel("Spending Score")
```

```python
# We can see that ages 0-40 have a very high spending score, only very few times they spend less money
# Ages from 40-70 have a low or average spending score ranging upto 60 only.
# Hence we can conclude that we need to target young ones for marketing purpose
```

```python
plt.scatter(X[:,0], X[:,1], c = kmeans.labels_, cmap = "rainbow")
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], color = "black")

plt.title("Plot between Age and Spending Score")
plt.xlabel("Age")
plt.ylabel("Spending Score")
```
